# Face & Eyes Detection In Images

**Import the required packages**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

**Load Face and Eye XML file using cascade classifier **

```python
# The cascade classifiers are the trained.xml files for detecting the face and eyes.
faceCascade=cv2.CascadeClassifier("HaarCascadeClassifier/haarcascade_frontalface_default.xml")
eyeCascade=cv2.CascadeClassifier("HaarCascadeClassifier/haarcascade_eye.xml")
```

**Load an image using 'imread' specifying the path to image**

```python
inputImage = cv2.imread("AlbertEinstein.png")
```

**Print each dimension of the image**

```python
print("Height of the input image is %d pixels"%(inputImage.shape[0]))
print("Width of the input image is %d pixels"%(inputImage.shape[1]))
```

**Display the input image**

```python
plt.imshow(cv2.cvtColor(inputImage, cv2.COLOR_BGR2RGB))
```

```python
# Our classifier returns the ROI of the detected face as a tuple
# It stores the top left coordinate and the bottom right coordiantes
faceDetect = faceCascade.detectMultiScale(inputImage,1.1,5)
print(faceDetect)
```

```python
# When no faces detected, face_classifier returns and empty tuple
if faceDetect is ():
    print('No faces found')
```

```python
# We iterate through the faceDetect array and draw a rectangle
for(x,y,w,h) in faceDetect:
    cv2.rectangle(inputImage,(x,y),(x+w,y+h),(255,127,200),10)
    roi = inputImage[y:y+h, x:x+w]
    eyesDetect = eyeCascade.detectMultiScale(roi)
    
    # We iterate through the eyesDetect array and draw a rectangle
    for (ex,ey,ew,eh) in eyesDetect:
        cv2.rectangle(roi,(ex,ey),(ex+ew,ey+eh),(255,255,0),5)
        plt.imshow(cv2.cvtColor(inputImage, cv2.COLOR_BGR2RGB))
        
        #To save the final output image use 'imwrite' specificing the file name and the image to be saved
        cv2.imwrite("Face&EyeDetectedImage.png",inputImage)
```
